## 4) Data Type
### 4.1) Numbers

- `int` จำนวนเต็ม เช่น `10`, `-3`
- `float` ทศนิยม เช่น `3.14`, `-0.5`
- `complex` จำนวนเชิงซ้อน เช่น `1 + 2j`

แปลงชนิดข้อมูลด้วย `int()`, `float()`, `complex()`


````python
x = 10        # Integer
y = 20.5      # Float
z = 3 + 4j    # Complex

print("x:", x, type(x))
print("y:", y, type(y))
print("z:", z, type(z))

print(int(x + y))        # แปลงเป็นจำนวนเต็ม
print(float(x * y))      # แปลงเป็นจำนวนทศนิยม
print(complex(x + y))    # แปลงเป็นจำนวนเชิงซ้อน
print(complex(x + z))    # รวมกับ complex
````

### 4.2) String (ข้อความ)

`str` คือข้อมูลประเภทข้อความ เช่น `"Hello, World!"`
- รวมข้อความด้วย `+`
- แปลงตัวเลขเป็นข้อความด้วย `str()`


````python
x = 10   # Integer
y = 20.5 # Float
z = 3 + 4j

message = "ดอกกระเจียวบาน อีกไม่นานก็คงลา อ้ายหรือเจ้าอยู่ต่อเก่า...."

print(message)
print(str(x) + str(y))  # แปลงเป็นข้อความแล้วค่อยต่อกัน
````

### 4.3) Boolean (ค่าตรรกะ)

### 4.4) List (รายการ)

`list` ใช้เก็บข้อมูลหลายตัวในตัวแปรเดียว

## 5) Input (รับค่าจากผู้ใช้)

`input()` จะรับค่ามาเป็น **ข้อความ (str)** ก่อนเสมอ

### 5.1) แปลงชนิดข้อมูล (Type Conversion

เพราะ `input()` ได้ `str` เสมอ ถ้าจะเอาไปคำนวณต้องแปลงก่อน เช่น `int(...)` หรือ `float(...)`

ด้านล่างมี 2 แบบ:
1) แปลงตอนเก็บตัวแปร
2) แปลงตั้งแต่ตอน `input()`



````python
# แบบที่ 1: แปลงตอนเก็บตัวแปร (แยก 2 บรรทัด)

years1 = input("Years: ")
print("Your age is:", 2569 - int(years1))
print("type(years1):", type(years1))

````

````python
# แบบที่ 2: แปลงตั้งแต่ตอน input() (บรรทัดเดียว)

years2 = int(input("Years: "))
print("Your age is:", 2569 - years2)
print("type(years2):", type(years2))
````

````python
#ตัวอย่างเพิ่มเติม

age = int(input("Enter your age: "))
print("Your age is:", age)
print("type(age):", type(age))

price = float(input("Enter the price: "))
print("The price is:", price)
print("type(price):", type(price))
````

## 6) Arithmetic Operators (เครื่องหมายคำนวณ)

ตัวอย่างเครื่องหมายคำนวณที่ใช้บ่อย: `+ - * / % ** //`


````python
x = 10
y = 3

print("x + y =", (x + y))   
print("x - y =", (x - y))   
print("x * y =", (x * y))   
print("x / y =", (x / y))   
print("x % y =", (x % y))   
print("x ** y =", (x ** y)) 
print("x // y =", (x // y)) 
````

### 6.1) String Operator (ใช้เครื่องหมายกับข้อความ)

ใช้ได้หลักๆ 2 แบบ:
- `+` ต่อข้อความ (concatenate)
- `*` ทำซ้ำข้อความ (repeat)


````python
# การบวก string (ต่อข้อความ)

str1 = "Hello"
str2 = "World"
result_str_add = str1 + " " + str2  # ผลลัพธ์คือ "Hello World"
print(result_str_add)
````

````python
# การคูณ string (การทำซ้ำ)

result_str_mul = str1 * 3  # ผลลัพธ์คือ "HelloHelloHello"
print(result_str_mul)
````

### 6.2) List กับเครื่องหมาย `+` และ `*`

`list` รองรับบาง operator:
- `+` เอา list มาต่อกัน (concatenate)
- `*` ทำซ้ำ list

แต่ `list` **ไม่สามารถ** ใช้ `-` หรือ `/` แบบตัวเลขได้โดยตรง


````python
# ตัวอย่างการบวก list

list1 = [1, 2, 3]
list2 = [4, 5, 6]

result_add = list1 + list2  # ผลลัพธ์คือ [1, 2, 3, 4, 5, 6]
print(result_add)
````

````python
# ตัวอย่างการคูณ list (ทำซ้ำ)

list3 = [1, 2, 3]
result_mul = list3 * 3      # ผลลัพธ์คือ [1, 2, 3, 1, 2, 3, 1, 2, 3]
print(result_mul)
````

## 7) Comparison Operators (ตัวเปรียบเทียบ)

ใช้เปรียบเทียบค่า แล้วจะได้ผลลัพธ์เป็น `True/False`
- `==` เท่ากับ
- `!=` ไม่เท่ากับ
- `>` มากกว่า, `<` น้อยกว่า
- `>=` มากกว่าหรือเท่ากับ, `<=` น้อยกว่าหรือเท่ากับ


````python
x = 10
y = 5

print(x == y)  
print(x != y) 
print(x > y)   
print(x < y)   
print(x >= y)  
print(x <= y)  
````

## 8) Logical Operators (ตัวดำเนินการทางตรรกะ)

- `and` เป็นจริงเมื่อทั้งสองข้างเป็นจริง
- `or` เป็นจริงเมื่ออย่างน้อยหนึ่งข้างเป็นจริง
- `not` กลับค่าจากจริงเป็นเท็จ/เท็จเป็นจริง


````python
a = True
b = False

print(a and b)  
print(a or b)   
print(not a)    
````

## 9) if / elif / else

ใช้เมื่ออยากให้โปรแกรม “ตัดสินใจ” ตามเงื่อนไข

โครงสร้างพื้นฐาน:
```python
if เงื่อนไข:
    # คำสั่งเมื่อเงื่อนไขเป็นจริง
elif เงื่อนไข:
    # คำสั่งเมื่อเงื่อนไขเป็นจริง (เงื่อนไขถัดมา)
else:
    # คำสั่งเมื่อทุกเงื่อนไขเป็นเท็จ
```


````python
# ตัวอย่าง if-else: รับคะแนน ถ้า >= 5 ให้ผ่าน

user_input = int(input("Input your score: "))

if user_input >= 5:
    print("Pass!")
else:
    print("Fail!")
````

### เฉลย (อย่าเเอบดูนะ)

```python

if 15 <= temp <= 25:
    print("สตรอว์เบอร์รี่")
```
