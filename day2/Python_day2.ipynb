# DataType (ชนิดข้อมูล)

Day 2 จะเจาะลึกการใช้งาน `tuple`, `set`, `dict` และ `list` พร้อมเมธอดที่ใช้บ่อย


## TUPLE

`tuple` คล้าย list แต่ **แก้ไขค่าไม่ได้** หลังจากสร้างแล้ว
- เหมาะสำหรับข้อมูลที่ไม่ควรเปลี่ยน เช่น พิกัดตำแหน่ง
- เข้าถึงข้อมูลด้วย index เช่น `coordinates[0]`


```python
coordinates = (10.0, 20.0)
bangkok_location = (13.7563, 100.5018)

print(coordinates[0])
```

## SET

`set` เก็บข้อมูลที่ **ไม่ซ้ำกัน** และ **ไม่เรียงลำดับ**
- `.add(value)` เพิ่มข้อมูล
- `.remove(value)` ลบข้อมูล


```python
unique_numbers = {1, 2, 3, 4, 5} # Set

print("Add")
print(f"Before: {unique_numbers}")

# เพิ่มข้อมูล
unique_numbers.add(6)

print(f"After: {unique_numbers}")

#######
print("\n================================\n")
#######

print("Remove")
print(f"Before: {unique_numbers}")

unique_numbers.remove(3)


print(f"After: {unique_numbers}")
```

## DICTIONARY

`dict` เก็บข้อมูลแบบคู่ `key: value`
- ใช้ `dict[key]` เพื่อดึงค่า
- เพิ่ม key ใหม่ได้ด้วย `dict['new_key'] = value`


```python
person = {
    "Name": "Phai",
    "Age": 35,
    "City": "Bangkok"
}

print(f"Key['Name'] | value = {person['Name']}")
print(f"Key['Age']  | value = {person['Age']}")
print(f"Key['City'] | value = {person['City']}")
```

```python
# value ใน dict สามารถเป็นชนิดข้อมูลใดก็ได้ เช่น str, int, bool

data = {
    "Name": "John",
    "Age": 19,
    "Is_happy": True
}

print(f"Key['Name']     | value = {data['Name']}")
print(f"Key['Age']      | value = {data['Age']}")
print(f"Key['Is_happy'] | value = {data['Is_happy']}")
```

```python
# dict ที่มี key เดียว

person = {
    "name": "Pai"
}

print(person['name'])
```

```python
# เพิ่ม key ใหม่เข้าไปใน dict ที่มีอยู่แล้ว

person = {
    "name": "Pai",
    "age": 35,
    "city": "Bangkok"
}

print("Add")
print(f"Before: {person}")

person['color'] = "red"

print(f"After: {person}")
```

## LIST

`list` ใช้เก็บข้อมูลหลายตัว แก้ไขได้ และเรียงลำดับ
- `.append(value)` เพิ่มที่ท้าย list
- `.remove(value)` ลบค่าที่ระบุ
- `list[index]` เข้าถึงหรือแก้ไขค่าตาม index


```python
fruits = ["apple", "papaya", "banana", "orange"]
number = [1, 2, 3, 4, 5]  # List

# Added
print("Add")
print(f"Before: {fruits}")

fruits.append("cherry")

print(f"After: {fruits}")

#######
print("\n================================\n")
#######

# Access
print(f"Index[1] = {fruits[1]}")

#######
print("\n================================\n")
#######

# Change
print(f"Change")
print(f"Before: {fruits}")

fruits[1] = "Demon fruit"

print(f"After: {fruits}")

#######
print("\n================================\n")
#######

print(f"remove")
print(f"Before: {fruits}")

fruits.remove("banana")

print(f"After: {fruits}")
```

# LOOP (การวนซ้ำ)

ใช้เมื่ออยากให้โปรแกรมทำงานซ้ำๆ โดยไม่ต้องเขียนโค้ดซ้ำหลายบรรทัด

## FOR LOOP

วนซ้ำตามจำนวนที่กำหนด โดยใช้ `range(start, stop, step)`
- `start` ค่าเริ่มต้น
- `stop` ค่าสิ้นสุด (ไม่รวมค่านี้)
- `step` ก้าวละเท่าไหร่


```python
for i in range(1, 11, 1):
    print("Hello world")
```

```python
for i in range(1, 11, 1):
    print("🍎")
```

```python
# step = 2 จะข้ามทีละ 2 ค่า → พิมพ์แค่เลขคู่

for i in range(0, 8, 2):
    print(i)
```

## WHILE LOOP

วนซ้ำตราบใดที่เงื่อนไขยังเป็น `True`
- ต้องระวัง **infinite loop** ถ้าเงื่อนไขไม่เคยเป็น `False`
- อย่าลืมอัพเดทตัวแปรใน loop เช่น `i += 1`


```python
i = 1

while i < 11:
    print(i)
    i += 1
```

# NESTED LOOP (loop ซ้อน loop)

วน loop ข้างในทุกรอบสำหรับแต่ละรอบของ loop ข้างนอก
- ใช้บ่อยกับข้อมูล 2 มิติ เช่น ตาราง หรือ matrix


```python
for i in range(0, 3, 1):
    for j in range(0, 3, 1):
        print(f"Hello World[{i}][{j}]")
```

## BREAK

`break` หยุด loop ทันทีเมื่อเงื่อนไขที่กำหนดเป็นจริง


```python
for i in range(0, 10, 1):
    if(i == 5):
        break

    print(i)
```

## CONTINUE

`continue` ข้ามรอบปัจจุบันแล้ววนต่อรอบถัดไปทันที
- ต่างจาก `break` ตรงที่ loop **ยังไม่หยุด** แค่ข้ามรอบนั้น


```python
for i in range(0, 10, 1):
    if(i == 5):
        continue

    print(i)
```

## LOOP STRING INDEX

วน loop ผ่านแต่ละตัวอักษรใน string ได้โดยตรง
- ตัวแปร `c` จะรับค่าตัวอักษรทีละตัวในแต่ละรอบ


```python
string = "PYTHON"
index = 0

for c in string:
    print(f"Index:{index} | Value:{c}")

    index += 1
```

## FUNCTION (ฟังก์ชัน)

ฟังก์ชันคือกลุ่มโค้ดที่ตั้งชื่อไว้ เรียกใช้ซ้ำได้ทุกที่
- ใช้ `def` เพื่อประกาศฟังก์ชัน
- รับ **parameter** เพื่อส่งค่าเข้าไป
- ใช้ `return` เพื่อส่งค่ากลับออกมา


```python
# ฟังก์ชันพื้นฐาน ไม่มี parameter และไม่มี return

def function_name():
    print("Hello")

function_name()
```

```python
# ฟังก์ชันที่รับ parameter เพื่อแสดงผลต่างกันตามค่าที่ส่งเข้ามา

def greet(name):
    print("Hello", name)

greet("John")
greet("Alice")
```

```python
# ฟังก์ชันที่รับ 2 parameter และ return ค่ากลับมา

def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```
